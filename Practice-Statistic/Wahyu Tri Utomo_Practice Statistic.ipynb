```python
#import dataset. 
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Syukrondzeko/Fellowship/main/Seasons_Stats.csv")
```

```python
#DATA PRE-PROCESSING
#IMPORT
#FILTER DATA
#HANDLING MISSING VALUE
#FILTERING COLUMN
```

```python
df.shape
```

```python
df.head()
```

```python
#Filtering data, to show data in 2017. 
df = df[df['Year'] == 2017]
df
```

```python
df.isnull().sum()
```

```python
#drop missing value that have as many missing values as the entire of row data
df.drop(['blanl', 'blank2'], axis=1, inplace=True)
df.isnull().sum()
```

```python
#delete duplicate values 
#keep the lastest one
df = df.drop_duplicates(subset=['Player'], keep='last')
df
```

```python
#
#
#NUMBER ONE
#
#
```

```python
#create df to answer question number 1
df_no1 = df[['Tm','Age', 'Player']]
df_no1
```

```python
#import numpy to use aggregate function
import numpy as np
```

```python
#Show the olders player for each team
df_no1.groupby('Tm').agg(np.max)
```

```python
#show youngest player for each team
df_no1.groupby('Tm').agg(np.min)
```

```python
#
#
#NUMBER TWO
#
#
```

```python
#create df to answer question number 2
df_no2 = df[['Player', 'Pos', 'MP']]
df_no2.head()
```

```python
#show the most minuted played in each position
df_no2.groupby('Pos').agg(np.max)
```

```python
#Create df to answer question number 3
df_no3 = df[['Tm', 'TRB%', 'AST%', 'STL%', 'BLK%']]
df_no3.head()
```

```python
#find team with highest average precentage of TRB%
a = df_no3.groupby('Tm').agg(np.mean)
a.sort_values(by=['TRB%'], ascending=False)
```

```python
#find team with highest average precentage of AST%
a = df_no3.groupby('Tm').agg(np.mean)
a.sort_values(by=['AST%'], ascending=False)
```

```python
#find team with highest average precentage of STL%
a = df_no3.groupby('Tm').agg(np.mean)
a.sort_values(by=['STL%'], ascending=False)
```

```python
#find team with highest average precentage of BLK%
a = df_no3.groupby('Tm').agg(np.mean)
a.sort_values(by=['BLK%'], ascending=False)
```
