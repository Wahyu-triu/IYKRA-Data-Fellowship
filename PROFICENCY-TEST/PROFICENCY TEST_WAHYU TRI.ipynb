```python
#merubah data categorik menjadi numerik
#menggunakan model desicion tree

#data bulan
data_model['MONTH'] = data_model['MONTH'].apply(lambda x: x.replace('Oktober', '10'))
data_model['MONTH'] = data_model['MONTH'].apply(lambda x: x.replace('November', '11'))
data_model['MONTH'] = data_model['MONTH'].apply(lambda x: x.replace('Deesember', '12'))
data_model['MONTH'] = data_model['MONTH'].apply(lambda x: x.replace('September', '9'))
data_model['MONTH'] = data_model['MONTH'].apply(lambda x: x.replace('Agustus', '8'))
data_model['MONTH'] = data_model['MONTH'].apply(lambda x: x.replace('Juli', '7'))
data_model['MONTH'] = data_model['MONTH'].apply(lambda x: x.replace('Juni', '6'))
data_model['MONTH'] = data_model['MONTH'].apply(lambda x: x.replace('Mei', '5'))
data_model['MONTH'] = data_model['MONTH'].apply(lambda x: x.replace('April', '4'))
data_model['MONTH'] = data_model['MONTH'].apply(lambda x: x.replace('Maret', '3'))
data_model['MONTH'] = data_model['MONTH'].apply(lambda x: x.replace('Februari', '2'))
data_model['MONTH'] = data_model['MONTH'].apply(lambda x: x.replace('Januari', '1'))
```

```python
data_model['MONTH'] = data_model['MONTH'].astype('int64')
```

```python
#ubdah hasi ke numerik
data_model['DAY_OF_WEEK'] = data_model['DAY_OF_WEEK'].apply(lambda x: x.replace('Monday', '1'))
data_model['DAY_OF_WEEK'] = data_model['DAY_OF_WEEK'].apply(lambda x: x.replace('Tuesday', '2'))
data_model['DAY_OF_WEEK'] = data_model['DAY_OF_WEEK'].apply(lambda x: x.replace('Wednesday', '3'))
data_model['DAY_OF_WEEK'] = data_model['DAY_OF_WEEK'].apply(lambda x: x.replace('Thursday', '4'))
data_model['DAY_OF_WEEK'] = data_model['DAY_OF_WEEK'].apply(lambda x: x.replace('Friday', '5'))
data_model['DAY_OF_WEEK'] = data_model['DAY_OF_WEEK'].apply(lambda x: x.replace('Saturday', '6'))
data_model['DAY_OF_WEEK'] = data_model['DAY_OF_WEEK'].apply(lambda x: x.replace('Sunday', '7'))
```

```python
data_model['DAY_OF_WEEK'] = data_model['DAY_OF_WEEK'].astype('int64')
```

```python
#ubah district jadi numerik
data_model['DISTRICT'] = data_model['DISTRICT'].apply(lambda x: x.replace('A1', '1'))
data_model['DISTRICT'] = data_model['DISTRICT'].apply(lambda x: x.replace('A7', '2'))
data_model['DISTRICT'] = data_model['DISTRICT'].apply(lambda x: x.replace('A15', '3'))
data_model['DISTRICT'] = data_model['DISTRICT'].apply(lambda x: x.replace('B2', '4'))
data_model['DISTRICT'] = data_model['DISTRICT'].apply(lambda x: x.replace('B3', '5'))
data_model['DISTRICT'] = data_model['DISTRICT'].apply(lambda x: x.replace('C6', '6'))
data_model['DISTRICT'] = data_model['DISTRICT'].apply(lambda x: x.replace('C11', '7'))
data_model['DISTRICT'] = data_model['DISTRICT'].apply(lambda x: x.replace('D4', '8'))
data_model['DISTRICT'] = data_model['DISTRICT'].apply(lambda x: x.replace('D14', '9'))
data_model['DISTRICT'] = data_model['DISTRICT'].apply(lambda x: x.replace('E5', '10'))
data_model['DISTRICT'] = data_model['DISTRICT'].apply(lambda x: x.replace('E13', '11'))
data_model['DISTRICT'] = data_model['DISTRICT'].apply(lambda x: x.replace('E18', '12'))
data_model['DISTRICT'] = data_model['DISTRICT'].apply(lambda x: x.replace('External', '13'))
```
