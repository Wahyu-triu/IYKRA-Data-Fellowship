# Pre-Processing or Number 1
Make sure that the data is clean

```python
#importing module

import pandas as pd
import numpy as np

#read data
data = pd.read_csv('https://raw.githubusercontent.com/Syukrondzeko/Fellowship/main/epl_1819.csv')
data.head()
```

```python
#check datatype in ecah column
data.info()
```

```python
#change datatype attack passes
#delete',' before chage the dtype to int.
for i, item in data['attack_passes'].items():
    data['attack_passes'] = item.replace(',','')

#merubah tipe data
data['attack_passes'] = data['attack_passes'].astype('int64')
```

```python
#change data type attack passes long
#delete',' before chage the dtype to int.
for i, item in data['attack_passes_long'].items():
    data['attack_passes_long'] = item.replace(',','')

#merubah tipe data
data['attack_passes_long'] = data['attack_passes_long'].astype('int64')
```

```python
#change data type attack passes back
##delete',' before chage the dtype to int.
for i, item in data['attack_passes_back'].items():
    data['attack_passes_back'] = item.replace(',','')

#merubah tipe data
data['attack_passes_back'] = data['attack_passes_back'].astype('int64')
```

```python
#change data type defence_clearances
#delete',' before chage the dtype to int.
for i, item in data['defence_clearances'].items():
    data['defence_clearances'] = item.replace(',','')

#merubah tipe data
data['defence_clearances'] = data['defence_clearances'].astype('int64')
```

```python
#re-check data type
data.info()
```

```python
#cek summary data
data.describe()
```

```python
#check missing values
data.isnull().sum()
```

# Number 3

```python
#number 3
#make a subset from used column
data_3 = data[['Team', 'general_won', 'general_draw', 'general_points', 'attack_passes',
               'attack_passes_through', 'attack_passes_long', 'attack_passes_back',
              'attack_crosses', 'attack_corners_taken', 'attack_shots', 'attack_shots_on_target',
              'attack_goals_headed', 'attack_goals_penalty', 'attack_goals_box',
              'attack_goals_outsidebox', 'attack_goals_counter', 'attack_goals_freekick',
              'attack_posession', 'attack_pass_accuracy']]
data_3
```

```python
#normalize data value
from sklearn import preprocessing

#make it into array format using .values
data_3_values = data_3.values

#data which will be normalized
data_3_values_norm = data_3_values[:, 1:]
```

```python
#data normalization
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(0,1))

data_3_norm = min_max_scaler.fit_transform(data_3_values_norm)
```

```python
#make a new data set from normalize data
data_3_new = pd.DataFrame({'Team':data_3_values[:,0], 'general_won':data_3_norm[:,0], 
                           'general_draw':data_3_norm[:,1], 'general_points':data_3_norm[:,2], 
                           'attack_passes':data_3_norm[:,3], 'attack_passes_through':data_3_norm[:,4],
                           'attack_passes_long':data_3_norm[:,5], 'attack_passes_back':data_3_norm[:,6],
                           'attack_crosses':data_3_norm[:,7], 'attack_corners_taken':data_3_norm[:,8],
                           'attack_shots':data_3_norm[:,9], 'attack_shots_on_target':data_3_norm[:,10],
                           'attack_goals_headed':data_3_norm[:,11], 'attack_goals_penalty':data_3_norm[:,12],
                           'attack_goals_box':data_3_norm[:,13],'attack_goals_outsidebox':data_3_norm[:,14],
                           'attack_goals_counter':data_3_norm[:,15], 'attack_goals_freekick':data_3_norm[:,16],
                           'attack_posession':data_3_norm[:,17], 'attack_pass_accuracy':data_3_norm[:,18]
                          })
data_3_new
```

```python
x = data_3_new.mean(axis=1, numeric_only=True)
data_3_new['Average score'] = x
data_3_new.head()
```

```python
#looking for the team with best attack aspect. 
d = data_3_new[['Team', 'Average score']].head()
d.sort_values(by=['Average score'], ascending=False)
```

```python
#with with best posession, addition to analyze
c = data_3[['Team','attack_posession']].head()
c.sort_values(by=['attack_posession'], ascending=False)
```
